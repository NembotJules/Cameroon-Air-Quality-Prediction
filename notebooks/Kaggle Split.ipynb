```python
# Number of rows in the dataset
total_rows = len(test_target_value)

# Determine 20% of the total rows for 'Public' and the rest for 'Private'
public_size = int(0.2 * total_rows)
private_size = total_rows - public_size

# Create an array with 80% 'Private' and 20% 'Public'
usage = ['Public'] * public_size + ['Private'] * private_size

# Shuffle the 'Usage' list to randomly distribute 'Public' and 'Private'
np.random.shuffle(usage)

# Add the 'Usage' column to the dataset
test_target_value['Usage'] = usage
```
